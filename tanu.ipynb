```python
inp=data.drop(['instant','cnt','dteday'],axis=1)
out=data[['cnt']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inp,out,test_size = 0.20,random_state=12,shuffle=True)
```
